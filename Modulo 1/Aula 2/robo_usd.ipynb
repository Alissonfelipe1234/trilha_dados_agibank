```python
import requests
import time
from datetime import datetime

# --- CONFIGURA칂칏ES ---
MOEDA_ORIGEM = "USD"
MOEDA_DESTINO = "BRL"
LIMITE_ALERTA = 6.00
ARQUIVO_SAIDA = "cotacoes.txt" # Arquivo de texto simples

print("Iniciando o Rob칪 de Monitoramento...")

while True:
    url = f"https://economia.awesomeapi.com.br/last/{MOEDA_ORIGEM}-{MOEDA_DESTINO}"
    
    try:
        resposta = requests.get(url)
        
        if resposta.status_code == 200:
            dados = resposta.json()
            cotacao_atual = float(dados['USDBRL']['bid'])
            
            print(f"Cota칞칚o capturada: R$ {cotacao_atual:.2f}")
            
            # --- SALVANDO EM ARQUIVO DE TEXTO (SIMPLES) ---
            # Montamos uma linha de texto com Data - Valor
            data_hora = datetime.now().strftime('%d/%m/%Y %H:%M:%S')
            linha = f"{data_hora} | {MOEDA_ORIGEM}->{MOEDA_DESTINO} | R$ {cotacao_atual:.2f}\n"
            
            # 'a' significa append (adicionar ao final sem apagar o que j치 existe)
            with open(ARQUIVO_SAIDA, 'a') as arquivo:
                arquivo.write(linha)
            
            print(f"-> Salvo no arquivo {ARQUIVO_SAIDA}")
            
            # Regra de Neg칩cio (Alerta)
            if cotacao_atual > LIMITE_ALERTA:
                print("ALERTA: Cota칞칚o acima do limite!")
        else:
            print("Erro ao conectar na API.")
            
    except Exception as e:
        print(f"Erro: {e}")
        
    time.sleep(10) # Espera 10 segundos
```
