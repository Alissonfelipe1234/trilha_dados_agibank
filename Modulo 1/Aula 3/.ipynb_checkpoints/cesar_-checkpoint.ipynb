```python
LISTA_ALFABETO = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
def criptografar(texto, chave):
    texto_criptografado = ''
    for letra in texto:
        if letra in LISTA_ALFABETO:
            novo_index = (LISTA_ALFABETO.index(letra) + chave) % 26
            texto_criptografado = texto_criptografado + LISTA_ALFABETO[novo_index]
        else:
            texto_criptografado = texto_criptografado + letra
        
    return texto_criptografado

criptografar("trilha python?", 5)
```

```python
# Solução sala
LISTA_ALFABETO = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']

def criptografar(texto, chave):
    return cesar(texto, chave)

def descriptografar(texto, chave):
    chave = -chave
    return cesar(texto, chave)
    
def cesar(texto, chave):
    texto_criptografado = ''
    for letra in texto:
        if letra in LISTA_ALFABETO:
            novo_index = (LISTA_ALFABETO.index(letra) + chave) % 26
            texto_criptografado = texto_criptografado + LISTA_ALFABETO[novo_index]
        else:
            texto_criptografado = texto_criptografado + letra
        
    return texto_criptografado

chave = 5
mensagem = criptografar("trilha python?", chave)
print ("criptografada: ", mensagem)

mensagem = descriptografar(mensagem, chave)
print(mensagem)
```
