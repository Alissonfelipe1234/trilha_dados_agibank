```python
import socket
import threading
import sys
```

```python
# Configurações globais
# 0.0.0.0 permite que o servidor aceite conexões de qualquer IP na rede
HOST_SERVER = '0.0.0.0' 
PORTA = 5556  # Uma porta alta para não conflitar com o sistema
```

```python
sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
```

```python
sock.bind((HOST_SERVER, PORTA))
sock.listen(10) # Aceita apenas 1 conexão por vez neste exemplo simples
```

```python
s_teste = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
s_teste.connect(("8.8.8.8", 80))
s_teste.getsockname()[0]
```

```python
def iniciar_chat():
    print("=== CHAT P2P PYTHON ===")
    print("Para conectar, vocês precisam estar na mesma rede (Wi-Fi).")
    
    nome = input("Seu nome: ")
    escolha = input("Você quer ser (1) Servidor [Anfitrião] ou (2) Cliente [Convidado]? ")

    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    conexao = None
    
    if escolha == '1':
        # SERVIDOR: Fica esperando alguém conectar
        try:
            sock.bind((HOST_SERVER, PORTA))
            sock.listen(10) # Aceita apenas 1 conexão por vez neste exemplo simples
            
            # Mostra o IP do computador para facilitar
            # Truque para descobrir o IP da sua rede (Wi-Fi)
            # Tenta conectar no Google DNS (não envia dados de verdade) para o sistema dizer qual é nosso IP
            s_teste = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
            try:
                s_teste.connect(("8.8.8.8", 80))
                meu_ip = s_teste.getsockname()[0]
            except Exception:
                meu_ip = "127.0.0.1"
            finally:
                s_teste.close()
            print(f"\nSala criada! Seu IP é: {meu_ip}")
            print(f"Peça para o colega conectar neste IP.")
            print("Aguardando conexão...")
            
            conexao, endereco = sock.accept() # bloqueante
            print(f"\nConectado com {endereco}!")
            
        except Exception as e:
            print(f"Erro ao criar servidor: {e}")
            return
            
    elif escolha == '2':
        # CLIENTE: Conecta em quem criou a sala
        ip_destino = input("Digite o IP do Servidor (ex: 192.168.1.5): ")
        try:
            print("Tentando conectar...")
            sock.connect((ip_destino, PORTA)) # bloqueante ou except
            conexao = sock
            print("\nConectado com sucesso!")
        except Exception as e:
            print(f"Não foi possível conectar: {e}")
            return
    else:
        print("Opção inválida.")
        return
    # --- AQUI COMEÇA A MÁGICA DAS THREADS ---
    
    # Criamos a thread apenas para RECEBER mensagens
    # Ela roda a função 'receber_mensagens' em paralelo
    thread_receber = threading.Thread(
        target=receber_mensagens, 
        args=(conexao, "Parceiro")
    )
    # daemon=True faz a thread morrer quando o programa principal fechar
    thread_receber.daemon = True 
    thread_receber.start()

    # O programa principal fica livre para cuidar apenas de ENVIAR
    while True:
        try:
            msg = input("Você: ")
            if msg.lower() == 'sair':
                break
            conexao.send(msg.encode('utf-8'))
        except KeyboardInterrupt:
            break
            
    print("\nEncerrando chat...")
    conexao.close()
```

```python
def receber_mensagens(conexao, nome_parceiro):
    """
    Esta função vai rodar em SEGUNDO PLANO (numa Thread separada).
    Ela fica num loop infinito apenas escutando o que chega.
    """
    while True:
        try:
            msg_bytes = conexao.recv(1024) # bloqueante
            
            if not msg_bytes:
                print(f"\n[INFO] {nome_parceiro} desconectou.")
                conexao.close()
                sys.exit() # Encerra o programa or break
                
            mensagem = msg_bytes.decode('utf-8')
            
            print(f"\n[{nome_parceiro}]: {mensagem}")
            
            # Força o prompt a aparecer de novo na linha de baixo
            print("Você: ", end="", flush=True)
            
        except Exception as e:
            print(f"\n[Erro na conexão]: {e}")
            break
```

```python
iniciar_chat()
```
