```python
# Função auxiliar para limpar texto (remove acentos e põe em maiúsculo)
def limpar_texto(texto):
    if pd.isna(texto): return ""
    # Normaliza para remover acentos (ex: ã -> a)
    nfkd = unicodedata.normalize('NFKD', str(texto))
    texto_sem_acento = u"".join([c for c in nfkd if not unicodedata.combining(c)])
    # Maiúsculo e remove espaços nas pontas
    return texto_sem_acento.upper().strip()

# --- 1. Preparação dos dados (como antes) ---
devedores = pd.read_csv('devedores.csv', sep=';')
aposentadorias = pd.read_csv('aposentadorias.csv', sep=';', skiprows=2)

# Limpeza dos CPFs (apenas números)
devedores['cpf_limpo'] = devedores['CPF/CNPJ DEVEDOR'].astype(str).str.replace(r'\D', '', regex=True)
aposentadorias['cpf_limpo'] = aposentadorias['CPF'].astype(str).str.replace(r'\D', '', regex=True)

# --- 2. Limpeza dos Nomes (Novo Passo) ---
# Aplicamos a função de limpeza nas colunas de nome específicas
devedores['nome_limpo'] = devedores['NOME DO DEVEDOR'].apply(limpar_texto)
aposentadorias['nome_limpo'] = aposentadorias['Nome'].apply(limpar_texto)

# --- 3. O Cruzamento Duplo ---
# Agora a chave de ligação usa DUAS colunas: CPF e NOME
resultado = pd.merge(
    devedores, 
    aposentadorias, 
    # A lista 'left_on' são as chaves do 1º df, 'right_on' as do 2º df
    left_on=['cpf_limpo', 'nome_limpo'], 
    right_on=['cpf_limpo', 'nome_limpo'],
    how='inner'
)

# Selecionar colunas finais para visualização limpa
colunas_finais = ['NOME DO DEVEDOR', 'CPF/CNPJ DEVEDOR', 'Nome', 'CPF']
print(f"Correspondências confirmadas (CPF parcial + Nome exato): {len(resultado)}")
resultado[colunas_finais].head()
```

```python
# 1. Renomear para facilitar (nomes muito longos atrapalham a análise)
credito = credito.rename(columns={
    'SALDO DEVEDOR SEM HONORÁRIOS': 'Saldo',
    'SITUAÇÃO DO CRÉDITO': 'Situacao',
    'DATA DA INSCRIÇÃO': 'Data'
})

# 2. Converter dinheiro (String Brasileira) para Float (Padrão Americano)
# Remove o 'R$', remove o ponto de milhar, troca vírgula por ponto
credito['Saldo'] = credito['Saldo'].astype(str).str.replace('R$', '', regex=False) \
                         .str.replace('.', '', regex=False) \
                         .str.replace(',', '.', regex=False) \
                         .str.strip() #   Remove espaços em branco sobrando    

# Agora converte para número
credito['Saldo'] = pd.to_numeric(credito['Saldo'])

# 3. Converter Data (opcional, mas bom para ordenar)
# O dayfirst=True é vital para o formato brasileiro DD/MM/AAAA
credito['Data'] = pd.to_datetime(credito['Data'], dayfirst=True)
```

```python
# Agrupa por situação, soma os saldos e cria o gráfico
credito.groupby('Situacao')['Saldo'].sum().plot(
    kind='bar', 
    title='Total da Dívida por Situação',
    color=['orange', 'blue'],
    figsize=(8, 5),
    rot=0 # Deixa o texto do eixo X na horizontal
)
```
