```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
fig = px.bar(x=["a", "b", "c"], y=[1, 3, 2])
fig.show()
```

```python
df_bolsa = pd.read_csv('dados_bolsa.csv', parse_dates=['Data'])

fig = px.line(df_bolsa, 
              x='Data', 
              y=['Fechamento', 'Media_Movel_30d'],
              title='Agibank Ações',
              labels={
                  'value': 'Preço', 
                  'variable': 'Indicador'
            },
              color_discrete_sequence=['#2E86C1', '#E74C3C'])
fig.update_layout(xaxis=dict(rangeslider=dict(visible=True)))
fig.show()
```

```python
df_rh = pd.read_csv('dados_rh.csv')

fig = px.scatter(df_rh, 
                 x='Vendas', 
                 y='Comissao', 
                 color='Performance',
                 size='Vendas',
                 hover_name='Vendedor', # Título do tooltip
                 hover_data=['Gerente', 'Regiao'] # Dados extras que aparecem ao passar o mouse
                )
fig.show()
```

```python
fig.write_html("relatorio_vendas_interativo.html")
print("Arquivo HTML gerado! Pode abrir no navegador.")
```

```python
dfmapa = pd.read_csv('df_mapa.csv')

fig = px.scatter_map(dfmapa, 
                        lat="Latitude", 
                        lon="Longitude", 
                        color="Categoria",     # Segmentação por cor (igual o hue)
                        #size="Valor_Venda",    # Tamanho da bolinha (Ticket Médio)
                        hover_name="Categoria",
                        hover_data=["Valor_Venda"],
                        zoom=10, 
                        height=600)

# Configurando o estilo do mapa (OpenStreetMap é gratuito e não precisa de API Key)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(title="Dados de produtos - SP")

fig.show()
```

```python
dfmapa_moda = dfmapa[dfmapa['Categoria'] == 'Moda']
dfmapa_moda.shape
```

```python
fig = px.density_mapbox(dfmapa_moda, 
                        lat='Latitude', 
                        lon='Longitude', 
                        z='Valor_Venda', # Peso (Onde tem mais dinheiro?)
                        radius=25,       # Raio de influência do calor
                        zoom=10,
                        mapbox_style="open-street-map",
                        title="Mapa de Calor: Concentração de Faturamento")

fig.show()
```

```python
resumo = dfmapa.groupby('Categoria')['Valor_Venda'].sum().reset_index()

fig = px.bar(resumo, 
             x='Categoria', 
             y='Valor_Venda', 
             color='Categoria',
             title='Faturamento por Categoria',
             text_auto='.2s')

fig.show()


# testar mean, median, min, max, sum
```
